```python
import os
import os
import random
import shutil
```

```python

# 设置文件路径
file_path = 'ciml7404/yale32/'

# 遍历目录下的所有文件
for file_name in os.listdir(file_path):
    # 判断当前路径是否为文件夹
    if os.path.isdir(os.path.join(file_path, file_name)):
        continue
        
    if file_name ==".DS_Store":
        pass
    else:
        # 获取文件名和扩展名
        name, ext = os.path.splitext(file_name)

        # 用切片根据文件名判断文件类型
        # 使用了切片出的前五个字符来作为文件夹的名称，并将文件移动到对应的文件夹中。
        type_str = name[:9]
        folder_path = os.path.join(file_path, type_str)

        # 用split根据文件名判断文件类型
        # 使用split函数以符号“_”作为分隔符，将文件名分割成多个字符串，然后使用第一个字符串作为文件夹名称。
        # type_str = name.split("_")[0]
        # folder_path = os.path.join(file_path, type_str)

        # 如果文件夹不存在，则创建
        if not os.path.exists(folder_path):
            os.mkdir(folder_path)

        # 拼接文件路径
        file_path_new = os.path.join(folder_path, file_name)

       # 判断目标路径中是否已经存在同名文件
        if os.path.exists(file_path_new):
            print(f"File {file_name} already exists in folder {type_str}")
        else:
            # 移动文件到对应的文件夹中
            shutil.move(os.path.join(file_path, file_name), file_path_new)
            print(f"File {file_name} moved to folder {type_str}")
```

```python
def getFileList(dir, Filelist, ext=None):
    """
    获取文件夹及其子文件夹中文件列表
    输入 dir：文件夹根目录
    输入 ext: 扩展名
    返回： 文件路径列表
    """
    newDir = dir
    if os.path.isfile(dir):
        if ext is None:
            Filelist.append(dir)
        else:
            if ext in dir[-3:]:
                Filelist.append(dir)

    elif os.path.isdir(dir):
        for s in os.listdir(dir):
            newDir = os.path.join(dir, s)
            getFileList(newDir, Filelist, ext)

    return Filelist
```

```python
file_path = 'yale-32x32-ordered'

# i means the number of selected image for training per subject
for i in range(2,8):
    picknumber = i 
    # source folder path
    org_img_folder = file_path 
    # target folder path
    tar_img_folder = 'yale-train-test-111/yale'+ str(picknumber)+'-2'    
    img_format = 'png' 

    # go to the subfolder
    for root,dirs,files in os.walk(org_img_folder):
        if root != 'yale-32x32-ordered':           
            # get image list in each subjects' folder
            imglist = getFileList(root, [], img_format)  
            # get selected image list 
            samplelist = random.sample(imglist, picknumber)  
            print(imglist)
            print(samplelist)

            
            for imgpath in imglist:
                # replicate the selected image to training dataset
                if imgpath in samplelist:
                    shutil.copy(imgpath, tar_img_folder+'/train')
                
                # replicate not selected image to test dataset
                else:
                    print(imgpath)
                    shutil.copy(imgpath, tar_img_folder+'/test')

```
